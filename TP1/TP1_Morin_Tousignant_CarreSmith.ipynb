## Numéro 1

*Trouvez analytiquement $\pi(\lambda|X = \text{\textbf{x}})$ et déduisez la valeur de $f(\text{\textbf{x}})$. Indice : trouvez une fonction $g(\lambda)$ proportionnelle à la loi a posteriori, ce qui vous permettra de retrouver la même forme que la loi a priori $\pi(\lambda)$. Cette loi a priori $\pi(\lambda)$ est conjuguée à la vraisemblance $f(\text{\textbf{x}}|\Lambda=\lambda)$.*

Partons de la forme a postériori donnée, puis simplifions là:
\begin{align*}
    \pi(\lambda|X = \text{\textbf{x}}) &= \frac{\lambda^{\alpha-1}e^{-\beta \lambda}\beta^{\alpha}}{f(x)\Gamma(\alpha)}\prod\limits_{i=1}^T \lambda e^{-\lambda x_i} \\[.5cm]
    &= \frac{\lambda^{\alpha+T-1}e^{-\beta \lambda}\beta^{\alpha}}{f(x)\Gamma(\alpha)}e^{-\lambda \sum_{i=1}^T x_i} \\[.5cm]
    &= \frac{\beta^{\alpha}\lambda^{\alpha+T-1}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)}}{f(x)\Gamma(\alpha)}
\end{align*}

Posons $g(\lambda)$ ainsi:
$$ g(\lambda) = \frac{\beta^{\alpha}\lambda^{\alpha+T-1}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)}}{\Gamma(\alpha)}$$

Cette fonction est proportionnelle à la fonction à postériori, et est composé du conjugué de $\pi(\lambda)$ et de $f(x|\Lambda=\lambda)$. On sait que f(x) agit comme une constante de normalisation dans la loi a postériori, qui considère toute les valeurs possible de $\lambda$ du conjugué de la fonction a priori et de vraissamblance. Donc, pour trouver f(x), il suffit de trouver la valeur de $g(\lambda)$ intégré sur tous les $\lambda$, i.e

\begin{align*}
    f(x) &= \int_0^\infty g(\lambda)d\lambda \\
    &= \frac{\beta^{\alpha}}{\Gamma(\alpha)}\int_0^\infty \lambda^{\alpha+T-1}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)}d\lambda
\end{align*}

Cette forme est très similaire à une fonction $\Gamma$. Utilisons tout de même Wolfram Alpha pour résouder l'intégrale analytiquement:

$$f(x) = \frac{\beta^{\alpha}}{\Gamma(\alpha)}\left[\Gamma(\alpha+T)\left(\beta+\sum_{i=1}^T x_i\right)^{-\alpha-T}\right]$$

Ainsi, on peut trouver la forme finale de $\pi(\lambda|X = \text{\textbf{x}})$
\begin{align*}
    \pi(\lambda|X = \text{\textbf{x}}) &= \frac{\beta^{\alpha}\lambda^{\alpha+T-1}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)}}{f(x)\Gamma(\alpha)} \\[.5cm]
    \longrightarrow \pi(\lambda|X = \text{\textbf{x}}) &= \frac{\lambda^{\alpha+T-1}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)\left(\beta+\sum_{i=1}^T x_i\right)^{\alpha+T}}}{\Gamma(\alpha+T)}
\end{align*}


## Numéro 2

*Évaluez analytiquement l’espérance a priori $\lambda_0:=E[\Lambda]=\int_0^\infty \pi(\lambda)d\lambda$ et l’espérance a posteriori $\hat{\lambda}$. Rappel : $\Gamma(z):=\int_0^\infty t^{z-1}e^{-t}dt$ et $\Gamma(z+1)=z\Gamma(z)$.*

**Calcul de $\lambda_0$**
\begin{align*}
    \lambda_0 &= \frac{\beta^{\alpha}}{\Gamma(\alpha)}\int_0^\infty \lambda^{\alpha-1}e^{-\beta\lambda}d\lambda \\[.5cm]
    &= \frac{\beta^{\alpha}}{\Gamma(\alpha)}\left(\beta^{-\alpha}\Gamma(\alpha)\right) \\[.5cm]
    &= 1
\end{align*}

**Calcul de $\hat{\lambda}$**
\begin{align*}
    \hat{\lambda} &= \frac{\left(\beta+\sum_{i=1}^T x_i\right)^{\alpha+T}}{\Gamma(\alpha+T)}\int_0^\infty\lambda^{\alpha+T}e^{-\lambda\left(\beta+\sum_{i=1}^T x_i\right)} d\lambda \\[.6cm]
    &= \frac{\left(\beta+\sum_{i=1}^T x_i\right)^{\alpha+T}}{\Gamma(\alpha+T)}\left[\Gamma(\alpha+T+1)\left(\beta+\sum_{i=1}^T x_i\right)^{-\alpha-T-1}\right] \\[.6cm]
    &= \frac{(\alpha+T)\,\Gamma(\alpha+T)}{\Gamma(\alpha+T)\left(\beta+\sum_{i=1}^T x_i\right)} \\[.6cm]
    &= \frac{\alpha+T}{\beta+\sum_{i=1}^T x_i}
\end{align*}

## Définition d'une fonction d'intégration par la méthode de Romberg

```python
def displaymat(func: Callable, a: int, b: int, R: np.ndarray[np.float64]):
    """Display the Romberg result matrix for easier debugging.

    Parameters
    ----------
        func
            The function that has been integrated.
        a
            The lower limit of integration.
        b
            The upper limit of integration.
        R
            The Romberg result matrix.
    """
    # Initialize matrix indexes
    i = j = 0
    _len = len(R)

    # Compute the number of function evaluations
    nbf = 2**(_len - 1) + 1

    print(f"Romberg integration of {repr(func)} from {[a, b]} \n")
    print(f"Steps \t StepSize Results")
    idx = np.arange(_len)
    stepsize = (b - a) / 2**idx
    for i in idx:
        print("%d \t %d \t " % (2**i, stepsize[i]), end=" ")
        for j in range(i+1):
            print("%.3e" % (R[i, j]), end=" ")
        print("")
    print(f"\nThe final result is: {R[i, j]} after {nbf} function evaluations.")


def romb(
    f: Callable,
    a: int,
    b: int,
    n: int = 10,
    tol: float = 1e-6,
    show: bool = False,
    **kwargs,
) -> np.float64:
    """Compute the integral of 'f' over [a, b] using Romberg integration.

    Parameters
    ----------
        f
            The function to integrate.
        a
            The lower limit of integration. 
        b
            The upper limit of integration.
        n
            The number of iterations to use.
        tol
            The relative error tolerance.
        show
            Whether to print the results. Defaults to False.

    Returns
    -------
        intgr
            The approximate value of the integral.
        error
            The error resulting from the integration mathod.
    """
    # Initialize the Romberg integration table
    R = np.empty((n, n))

    # Configure function f to pass extra arguments
    f = partial(f, **kwargs)

    # Compute the trapezoidal rule for the first column
    h = b - a
    R[0, 0] = 0.5 * h * (f(a) + f(b))

    # Iterate to refine the approximation
    idx = np.arange(1, n)
    h *= 0.5**idx
    for i in idx:
        k = np.arange(1, 2**i, 2)
        sum_f = f(a + k * h[i-1]).sum()
        R[i, 0] = 0.5 * R[i-1, 0] + sum_f * h[i-1]

        # Compute the rest of the column using Richardson extrapolation
        for j in range(1, i+1):
            R[i, j] = R[i, j-1] + (R[i, j-1] - R[i-1, j-1]) / (4**j - 1)

        # Compute integration error
        intgr = R[i, i]
        if intgr == 0:
            error = np.inf
        else:
            error = np.abs((intgr - R[i-1, i-1]) / intgr)

        if error < tol:
            break

    # Send table to display function for debuging purposes if specified
    if show is True:
        displaymat(f, a, b, R)
    return intgr, error


def simpson(f: Callable,
    a: int,
    b: int,
    n: int = 10,
    **kwargs,
) -> np.float64:
    """Compute the integral of 'f' over [a, b] using Simpson integration.

    Parameters
    ----------
        f
            The function to integrate.
        a
            The lower limit of integration. 
        b
            The upper limit of integration.
        n
            The number of iterations to use.

    Returns
    -------
        intgr
            The approximate value of the integral.
        error
            The error resulting from the integration mathod.
    """
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x, **kwargs)

    intgr = h / 3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    error = (b - a) * ((h**4) / 180) * np.abs(y).max()
    return intgr, error
```

## Résolution du problème

```python
# Problem constants
ALPHA = 2
BETA = 0.25


def bayesian_inf(
    lamb: np.float64,
    xdata: np.ndarray[np.float64],
    fx: Optional[np.float64] = None,
) -> np.ndarray[np.float64]:
    """Compute the normalized or unnormalized posterior values.

    Parameters
    ----------
        lamb
            The rate parameter of the exponential distribution.
        xdata
            The time difference dataset of the distribution.
        fx
            The 'evidence' or 'Marginal likelihood' of the distribution.

    Returns
    -------
        post
            The normalized posterior values if 'fx' is given, the unnormalized
            value otherwise.
    """
    # Compute the prior
    prior = st.gamma.pdf(lamb, ALPHA, scale=1/BETA)

    # Compute the liklihood
    likelihood = np.prod(lamb * np.exp(-lamb * xdata))

    # Compute the distribution's posterior
    post = likelihood * prior
    if fx is not None:
        post *= lamb / fx
    return post

bayesian_inf_vectorize = np.vectorize(bayesian_inf, excluded=["xdata", "fx"])

def get_fx(
    data: int,
    a: int,
    b: int,
    n: Optional[int] = None,
    method: str = "simpson",
) -> np.float64:
    """Compute f(x) and MLE using Simpson's integration rule.

    Parameters
    ----------
        data
            The dataset to integrate using Simpson's rule.
        a
            The lower limit of integration.
        b
            The upper limit of integration.
        n
            The number of samples between the integration limits. Defaults to
            50, the default value of the 'linspace' function.
        method
            The integration method to use. Available options are 'simpson' and
            'romberg'. Defaults to 'simpson'.

    Returns
    -------
        fx
            The 'evidence' or 'Marginal likelihood' of the distribution.
        lamb_hat
            The Maximum Likelihood Estimate (MLE) for lambda.

    Raises
    ------
    ValueError
        If the input method is nether 'simpson' nor 'romberg'.
    """
    if n is None:
        n = b + 100

    if method == "simpson":
        fx, fx_err = simpson(bayesian_inf_vectorize, a, b, n, xdata=data)
        lamb_hat, lamb_hat_err = simpson(bayesian_inf_vectorize, a, b, n, xdata=data, fx=fx)
    elif method == "romberg":
        fx, fx_err = romb(bayesian_inf_vectorize, a, b, n, xdata=data)
        lamb_hat, lamb_hat_err = romb(bayesian_inf_vectorize, a, b, n, xdata=data, fx=fx)
    else:
        raise ValueError("Available methods are 'simpson' and 'romberg'")
    return fx, fx_err, lamb_hat, lamb_hat_err


lamb_calc = np.empty(3)
for k in range(3):
    tdata = np.ediff1d(loadtxt(k))
    _simp = get_fx(tdata, 0, 200, 1000)
    _romb = get_fx(tdata, 0, 200, 15, "romberg")

    # Storing lambda values from Romberg's method for future use
    lamb_calc[k] = _romb[2]
    print(f"Results for dataset {k}:")
    print(f"Simpson: f(x) \u2248 {_simp[0]:.4e}, \u0394f(x) \u2248 {_simp[1]:.4e}, \u03BB_hat \u2248 {_simp[2]:.4f}, \u0394\u03BB_hat \u2248 {_simp[3]:.4e}")
    print(f"Romberg: f(x) \u2248 {_romb[0]:.4e}, \u0394f(x) \u2248 {_romb[1]:.4e}, \u03BB_hat \u2248 {_romb[2]:.4f}, \u0394\u03BB_hat \u2248 {_romb[3]:.4e} \n")
```

# Numéro 5

En comparant à votre solution analytique, tracez l’erreur sur $f(\text{\textbf{x}})$ et $\hat{\lambda}$ engendrée par vos deux méthodes d’intégration en fonction du nombre de tranches pour le jeux de données de votre choix.

```python
# Définir les valeurs analytiques de f(x) et lambda_hat
tdata = np.ediff1d(loadtxt(0))


analytical_fx = (BETA**ALPHA/gamma(ALPHA))*(gamma(ALPHA + (len(tdata))) * (BETA + np.sum(tdata))**(-ALPHA-(len(tdata))))
analytical_lambda = (ALPHA + len(tdata))/(BETA + np.sum(tdata))

# Paramètres pour le tracé
num_tranches = np.arange(2, 250, 2)  # Nombre de tranches à tester

# Calculer les erreurs pour la méthode de Simpson
errors_fx_simp = []
errors_lambda_simp = []
for n in num_tranches:
    fx, fx_err, lamb_hat, lamb_hat_err = get_fx(tdata, 0, 200, n, method="simpson")
    error_fx = np.abs(analytical_fx - fx)
    error_lambda = np.abs(analytical_lambda - lamb_hat)
    errors_fx_simp.append(error_fx)
    errors_lambda_simp.append(error_lambda)

# Calculer les erreurs pour la méthode de Romberg
errors_fx_romb = []
errors_lambda_romb = []
for n in num_tranches:
    fx, fx_err, lamb_hat, lamb_hat_err = get_fx(tdata, 0, 200, n, method="romberg")
    error_fx = np.abs(analytical_fx - fx)
    error_lambda = np.abs(analytical_lambda - lamb_hat)
    errors_fx_romb.append(error_fx)
    errors_lambda_romb.append(error_lambda)

# Tracé des erreurs
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(num_tranches, errors_fx_simp, label='Simpson')
plt.plot(num_tranches, errors_fx_romb, label='Romberg')
plt.xlabel('Nombre de tranches')
plt.ylabel('Erreur sur f(x)')
plt.title('Erreur sur f(x) en fonction du nombre de tranches')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(num_tranches, errors_lambda_simp, label='Simpson')
plt.plot(num_tranches, errors_lambda_romb, label='Romberg')
plt.xlabel('Nombre de tranches')
plt.ylabel('Erreur sur $\hat{\lambda}$')
plt.title('Erreur sur $\hat{\lambda}$ en fonction du nombre de tranches')
plt.legend()

plt.tight_layout()
plt.show()
```

Tracez un histogramme des fréquences relatives des écarts entre les impulsions. Tracez également la loi du modèle a priori $f(x_i|\Lambda = \lambda_0)$ et celle ajustée $f(x_i|\Lambda = \hat{\lambda})$.

```python
# Bin size: interval between two
# consecutive values in "ms"
NBINS = 30

# Initializing figures
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for k in range(3):
    t = np.ediff1d(loadtxt(k))

    # Generating distribution
    t_gen = np.linspace(0, t.max(), 1000)
    dist_prior = st.expon.pdf(t_gen, scale=t.mean())
    dist_posterior = st.expon.pdf(t_gen, scale=1/lamb_calc[k])

    # Creating plots
    ax = axs[k]
    ax.hist(t, NBINS, density=True, color="b", label="Frequency")
    ax.plot(t_gen, dist_prior, "r", lw=3, label="$prior$")
    ax.plot(t_gen, dist_posterior, "--g", lw=3, label="$posterior$")
    ax.set_xlabel("Time between samples [ms]")
    ax.set_ylabel("Relative frequency")
    ax.set_title(f"Dataset {k}")
    ax.legend()
```
